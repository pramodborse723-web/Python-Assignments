# 2. Tuple Manipulation

# Define tuple of 5 colors

```python
colors = ("Red", "Blue", "Green", "Yellow", "Black")
```

```python
print("Original Tuple:", colors)
```

# Tuples cannot be modified

```python
print("Tuples are immutable, so new items cannot be added directly.")
```

# Convert tuple to list

```python
color_list = list(colors)
```

# Add new color

```python
color_list.append("White")
```

# Convert back to tuple

```python
colors = tuple(color_list)
```

```python
print("Updated Tuple:", colors)
```

# 3. Set Operations

# Create two sets

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
```

```python
print("\nSet 1:", set1)
print("Set 2:", set2)
```

# Union

```python
print("Union:", set1.union(set2))
```

# Intersection

```python
print("Intersection:", set1.intersection(set2))
```

# Difference

```python
print("Difference:", set1.difference(set2))
```

# Remove one element and add new element

```python
set1.remove(2)
set1.add(10)
```

```python
print("Updated Set 1:", set1)
```
